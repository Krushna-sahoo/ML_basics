```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
data_set= pd.read_csv('train.csv')
print(data_set.shape)
```

```python
data_set.head()
```

```python
feature_With_na=[features for features in data_set.columns if data_set[features].isnull().sum()>1]

for features in feature_With_na:
    print(features,np.round(data_set[features].isnull().mean(),4),'% missing values')
```

```python
for features in feature_With_na:
    data= data_set.copy()

    data[features]=np.where(data[features].isnull(),1,0)
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
```

```python
#features which don't have any object or string values
numerical_values=[features for features in data_set.columns if data_set[features].dtype!='O']
data_set[numerical_values].head()
```

```python
year_feature=[features for features in numerical_values if 'Yr' in features or 'Year' in features]
year_feature
```

```python
for features in year_feature:
    print(features,data_set[features].unique())
```

```python
data_set.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.show()
```

```python
for features in year_feature:
    if features!='YrSold':
        data=data_set.copy()
        data[features]=data['YrSold']-data[features]
        plt.scatter(data[features],data['SalePrice'])
        plt.xlabel(features)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_values if len(data_set[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))

```

```python
discrete_feature
```

```python
data_set[discrete_feature]
```

```python
for features in discrete_feature:
    data=data_set.copy()
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.xlabel(features)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
continuous_feature=[feature for feature in numerical_values if feature not in discrete_feature+year_feature+['Id']]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

```python
data_set[continuous_feature].head()
```

```python
for feature in continuous_feature:
    data=data_set.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```
